# II. Data cleaning

We now need to prepare the data, meaning the following steps:
- Select relevant features
- Handle missing data, duplicates, outliers
- Preprocess the data (both categorical and quantitative)
- Possibly, make feature engineering

## II.1 Features selection

We might not need to keep all the given features, let's have a look at what we can drop:

````python
df.head()
````

Do you see columns that won't help predict the survival probability of a passenger? Or columns that might be difficult to handle properly?

Let's remove them.

````python
df = df.drop(columns=['PassengerId', 'Ticket'])
````

## II.2 Missing data

As previously stated, there is missing data:

````python
df.info()
````

How to deal with missing data?

There are several simplistic approaches:
- drop the entire column with `.drop()`: let's do that for the `'Cabin'` column
- drop the samples with missing data with `.dropna()`: let's do that for the `'Embarked'` column
- fill the data with `.fillna()`: let's do that for the `'Age'` column

````python
# Drop the Cabin column
df = df.drop(columns=['Cabin'])

# Drop the two lines with missing Embarked
df = df.dropna(subset=["Embarked"])

# Fill the data without age with mean age
df['Age'] = df['Age'].fillna(df['Age'].mean())
````

Let's check again our dataset with `.info()`: we now have 889 lines, and no missing data!

````python
df.info()
````

## II.3 Duplicates

One can check if there are duplicates in the data: be careful, some duplicates have to be kept!

````python
df.duplicated().sum()
````

We don't have duplicates here, but if we did we could have removed them with the method `.drop_duplicates()`

## II.4 Outliers

In some datasets, there can be outliers: data that do not make sense!

For example, broken sensor, misfilled information (a person have 220 years instead of 22...), and so on... Once again, not all outliers have to be removed!

# III. Data preparation

There are two main types of data:
- Categorical data: both binary and non-binary
- Continuous data

In our dataset, can you tell what columns belong to which category?

````python
df.head()
````

Those data have to be preprocessed differently. How would you preprocess those kind of data?

## III.1 Categorical data

Categorical data has to be one-hot-encoded: meaning every class has to be transformed into 0 and 1.

We can do that with the pandas function `pd.get_dummies()`

````python
pd.get_dummies(df['Sex']).head()
````

This transformed the strings `'male'` and `'female'` into `0` or `1`. 

But isn't there redundancy in the information here? We can remove one column automatically with the option `drop_first=True`:

````python
pd.get_dummies(df['Sex'], drop_first=True).head()
````

So that now we have well encoded our gender attribute to a binary variable.

What about non-binary categorical data?

How can we replace the `'Embarked'` column (being `C`, `S` or `Q`) with only 0 and 1?

````python
pd.get_dummies(df['Embarked']).tail()
````

Here, using again the function `get_dummies()`, we replaced `C`, `S` and `Q` with one-hot-encoding, meaning:
- `C` is now replaced by `(1, 0, 0)`
- `Q` is now replaced by `(0, 1, 0)`
- `S` is now replaced by `(0, 0, 1)`

Again, isn't this redundant? If this is not `C` nor `Q`, then this has to be `S`, right?

We can redo the `drop_first=True` trick to remove one columns and avoid redundancy:

````python
pd.get_dummies(df['Embarked'], drop_first=True).tail()
````

Now, we can preprocess all the categorical data in one line using the `get_dummies()` function. We first need to list all the categorical columns:

````python
categorical_columns = ['Sex', 'Embarked']
````

Then we can process them with `get_dummies()`:

````python
dummies = pd.get_dummies(df[categorical_columns], drop_first=True)
dummies.tail()
````

## III.2 Quantitative data

Now we can play with quantitative data.

Let's have a look at the range of the values:

````python
df.describe()
````

Some columns range from 1 to 3 (`Pclass`), while others climb up to 512 (`Fare`).

But we want to make sure all of those columns are considered to be at the same level: we need to rescale the data.

What strategies of rescaling can you imagine?

````python
quantitative_columns = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

scaled = (df[quantitative_columns] - df[quantitative_columns].mean())/df[quantitative_columns].std()

scaled.describe()
````

Now all our quantitative features are centered on 0 and have a standard deviation of 1.

## III.3 Feature Engineering

Sometimes, it might be useful to perform feature engineering: this is the art of creating new features!

In our example, we did not use at all the `Name` feature: how could we use it wisely?

````python
df['Name'].head(10)
````

We would keep the title for example, and see what are the existing titles in our dataset.

We can do that using the pandas method `.str` and `.split()`:

````python
df['Title']= df['Name'].str.split(' ').str[1]
df['Title'].value_counts()
````

We could split that into several classes: `Mr.`, `Miss.`, `Mrs.`, `Master.` and `Others` for example.

````python
df.loc[(df['Title']!='Mr.') & (df['Title']!='Miss.') & (df['Title']!='Mrs.') & (df['Title']!='Master.'), 'Title'] = 'Others'
df['Title'].value_counts()
````

We now have a new feature, the `Title`, which is a categorical class. What do we need to do before going further?

````python
titles = pd.get_dummies(df['Title'], drop_first=True)
titles.head()
````

## III.4 Concatenation

Finally, we need to concatenate all the information to make our dataset of features `X`. To do so, the pandas function `pd.concat()` can be used.

And we also need to create our target `y`: here this is the survival of the person.

````python
X = pd.concat([dummies, scaled, titles], axis=1)
X.head()
````

````python
y = df['Survived']
y.head()
````

## III.5 Splitting the data

It is particularly important to separate the data that you use for training your model, and the data that you use for testing your model.

It would be like a student practicing on sample questions for his/her exams, and being tested on the exact same question.

You need **to be tested on new questions** in order to make sure you learnt correctly.


We usually split it in a training set and a test set with the associated repartition: **80% - 20%**.

This convention allows to retrieve enough data to learn from it, and a small test subset in order to understand the performance.

To recap, we have:
- a **Training set** `(X_train, y_train)`: A set of observations used **for learning**. It fits the parameters (or weights) of the model.
- a **Test set** `(X_test, y_test)`: A set of observations (unseen during training) used only **to assess the performance** of the model.

## III.6 Saving the data

Because we may need to reuse a cleaned and prepared dataset without preprocessing it again, we may need to save the data.

A common tool for that in Python in `pickle`, it allows to save and load data in binary format really easily:

````python
import pickle

pickle.dump((X, y), open('clean_data.pkl', 'wb'))
````

We can also reload and reuse the data easily:

````python
load_X, load_Y = pickle.load(open('clean_data.pkl', 'rb'))
load_X.head()
````
